```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import time
import os
```

```python

url='https://costprojections.cancer.gov/expenditures.html'
tables =pd.read_html(url)
df= tables[0]
df.columns =["id","Base_2010","Base_2020","Trend_incidence_2020","Trend_survival__2020","Trend_incidence&survival_2020","2%_overall_2020","2%initial&lastyear_2020","5%initial&lastyear_2020"]
df
```

```python
new_url = 'https://www.verywellhealth.com/best-hospitals-for-cancer-treatment-514514'
response = requests.get(new_url)
```

```python
soup=bs(response.text,"html.parser")
```

```python
cancer_inst=[]

results = soup.find_all("div",class_="comp mntl-sc-list-item list-sc-item mntl-block")
for result in results:
    cancer_inst.append(result.h3.span.get_text())

cancer_df = pd.DataFrame(dict(institute = cancer_inst))
cancer_df
```

```python
csv_file ="C:/Users/gairo/Desktop/UTmygit/class-content/12-Web-Scraping-and-Document-Databases/cancer_inst.csv"
# path = os.getcwd()
# print(path)
latlng = pd.read_csv(csv_file,encoding="utf8")
latlng_df = pd.DataFrame(latlng)
frames = [cancer_df,latlng_df]
final_df = pd.concat(frames,axis = 1)
final_df
```
